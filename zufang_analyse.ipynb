```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'房价123.txt',header=None,names=['area','describe','room','square','price'])
df.info()
```

```python
### 增加每平米单价属性
square=df['square']
digital_square=square.str.replace('平米','').astype(int)
df['per_square_meter']=df['price']/digital_square
df
```

```python
### 区域租金分析
area_price_group=df.groupby('area')
## 租金和每平米租金查看
area_price=area_price_group.mean()
area_price
```

```python
### 按区域价格区间
area_price_dict={}
for area,price in area_price_group['price']:
    area_price_cut=pd.cut(price,bins=[0,1000,2000,3000,4000,100000]).value_counts()
    area_price_cut=area_price_cut.sort_index()
    area_price_dict[area]=area_price_cut
area_price_dict
    
```

```python
### 区域租金可视化
%matplotlib inline
## 解决中文乱码问题
import matplotlib as mpl
from matplotlib.font_manager import _rebuild
_rebuild()
mpl.rcParams['font.sans-serif']=[u'SimHei']
mpl.rcParams['axes.unicode_minus']=False

plt.bar(x=area_price.index,height=area_price['price'],)
```

```python
for key,item in area_price_dict.items():
    print(item.index.values)
    print(item.values)
```

```python
### 区域价格区间可视化
fig,axes=plt.subplots(3,3)
i=0
j=0
for key,item in area_price_dict.items():
    ax=axes[i,j]
    ax.set_title(key)
    ax.pie(item.values,labels=item.index.values,autopct='%1.1f',explode=[0.1,0,0,0,0.2])
    j+=1
    if j>2:
        i+=1
        j=0
```

```python
### 商家最爱用描述语分析
shorts=df['describe'].values.tolist()
## 分词
import jieba
texts_cut=jieba.lcut(''.join(shorts))
texts_cut
```

```python
##添加词典完善分词
for i in range(10):
    for j in range(4):
        jieba.add_word("{}室{}厅".format(i,j))
for i in range(1000,4000):
    jieba.add_word("{}元".format(i))
texts_cut=jieba.lcut(''.join(shorts))
texts_cut
```

```python
## 去除停用词
stopwords=[' ']
while True:
    try:
        for word in stopwords:
            texts_cut.remove(word)
    except:
        break
texts_cut
```

```python
## 词频统计
import jieba.analyse
words=jieba.analyse.textrank(''.join(shorts),topK=50,withWeight=True)
cloud_words=[word[0] for word in words]
```
