```python
import jieba
```

```python
text_list=datas['short'].tolist()
text_list
```

```python
text=';'.join(text_list)
text
```

```python
text_list=jieba.lcut(text)
texts=" ".join(text_list)
texts
```

```python
import jieba.analyse
import numpy as np
tfidf_word=jieba.analyse.extract_tags(texts,topK=100,withWeight=True)
textrank_word=jieba.analyse.textrank(texts,topK=100)
```

```python
import jieba.posseg as pseg
datas=pseg.cut("我爱北京天安门")
list(datas)
```

```python
list(tfidf_word)
```

```python
x,y=[],[]
for word in tfidf_word:
    x.append(word[0])
    y.append(word[1])
plt.bar(x[:50],y[:50])
```

```python
from wordcloud import WordCloud,ImageColorGenerator
font_path="/System/Library/fonts/PingFang.ttc"
wordcloud=WordCloud(font_path,background_color='white',width=1000,height=880).generate(" ".join(x))
plt.imshow(wordcloud)
```

```python
from PIL import Image
img_path="/Users/tudou/Downloads/bizhi/图片/2.jpg"
background = np.array(Image.open(img_path))
wordcloud=WordCloud(font_path,background_color='white',mask=background).generate(" ".join(x))
image_colors = ImageColorGenerator(background)
plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear")
```
